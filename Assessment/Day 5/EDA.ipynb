## 5. Outlier Detection

```python
sns.boxplot(y=mpg['mpg'])
plt.title('Box Plot for Total Bill')
plt.show()
```

## 6. Outlier Removal

```python
#Outlier Removal

Q1 = mpg['mpg'].quantile(0.25)
Q3 = mpg['mpg'].quantile(0.65)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outlier = mpg[(mpg['mpg'] < lower_bound) | (mpg['mpg'] > upper_bound)]
print(outlier)
#Remove outliers
mpg_clean = mpg[(mpg['mpg'] >= lower_bound) & (mpg['mpg'] <= upper_bound)]
mpg_clean
```

## 7. Correlation After Outlier Removal

```python
corr = mpg.corr(numeric_only=True)
print(corr)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```
